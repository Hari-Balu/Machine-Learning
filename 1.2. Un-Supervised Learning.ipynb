# Topics in un-Supervised Learning

# 1. Clustering

Clustering in unsupervised learning is a machine learning technique used to group similar data points together without using predefined labels or categories. The goal is to discover the inherent structure or patterns in a dataset.

✅ Key Concepts:

1. Unsupervised learning: Unlike supervised learning, where the model learns from labeled data, in clustering, the model learns from unlabeled data.

2. Cluster: A group of similar data points based on a defined similarity metric (e.g., Euclidean distance).

3. Objective: Maximize intra-cluster similarity (points in the same cluster are similar) and minimize inter-cluster similarity (points in different clusters are dissimilar).

📘 Common Clustering Algorithms:

1. K-Means Clustering

Partitions data into K clusters based on centroid proximity.

2. Hierarchical Clustering

Builds a tree (dendrogram) of clusters using agglomerative or divisive approaches.

3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Groups together points that are closely packed and marks outliers as noise.

4. Mean Shift

Finds clusters by updating centroids towards the mean of the data points in the neighborhood.

📌 Applications:

🔸Customer segmentation

🔸Image compression

🔸Market basket analysis

🔸Anomaly detection

🔸Document or text clustering

## 1.1. K-Means Clustering

K-Means Clustering is one of the most popular and straightforward clustering algorithms used in unsupervised learning. It partitions the dataset into K clusters, where each data point belongs to the cluster with the nearest mean (centroid).

```python
# Example 1: solve income group clustering problem

import os
# Fix memory leak warning on Windows with MKL and KMeans
os.environ["OMP_NUM_THREADS"] = "1"


import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler



# Step 1: Define the Dataset

df = pd.read_csv(r"C:\Users\admin\Downloads\income.csv")
print(df.head())

plt.scatter(df['Age'], df['Income($)'])


# Step 2: Initialize the parameter

x = df[['Age']]
y = df[['Income($)']]


# Step 3: Assign the cluster & Train the Data using K Means

KM = KMeans(n_clusters = 3)
KM.fit(df[['Age', 'Income($)']])


# Step 4: Predict the Dataset 

y_pred = KM.predict(df[['Age', 'Income($)']])
print('\n Y Predicted :', y_pred)

df['Clusters'] = y_pred
print('\n',df.head())

# Centeroid point of the Dataset

Centroid = KM.cluster_centers_
print('\n Centroid',Centroid)




# Step 5: Plot the Dataset

df1 = df[df['Clusters'] == 0]
df2 = df[df['Clusters'] == 1]
df3 = df[df['Clusters'] == 2]

plt.scatter(df1.Age, df1['Income($)'], color = 'red')
plt.scatter(df2.Age, df2['Income($)'], color = 'blue')
plt.scatter(df3.Age, df3['Income($)'], color = 'green')
plt.scatter(Centroid[:,0], Centroid[:,1], color = 'purple', marker = '+', label = 'Centroid')

plt.title('KMeans Clustering: Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income($)')

plt.legend()

plt.show()


# Step 6: Preprocessing - Use minmax scaler

Scaler = MinMaxScaler()
Scaler.fit(df[['Income($)']])
df['Income($)'] = Scaler.transform(df[['Income($)']])
print('\n',df.head())


Scaler.fit(df[['Age']])
df['Age'] = Scaler.transform(df[['Age']])
print('\n',df.head())



# Step 7: Train the Scaled Dataset 

KM = KMeans(n_clusters = 3)
KM.fit(df[['Age', 'Income($)']])


# Step 8: Predict the Scaled Dataset 

S_pred = KM.predict(df[['Age', 'Income($)']])
print('\n Predicted Scaled Dataset :', S_pred)

df['Scaled Clusters'] = S_pred
print('\n',df.head())


# Centeroid point of Scaled Dataset

S_Centroid = KM.cluster_centers_
print('\n Scaled Centroid',S_Centroid)

# Step 9: Plot the Scaled Dataset


df1 = df[df['Scaled Clusters'] == 0]
df2 = df[df['Scaled Clusters'] == 1]
df3 = df[df['Scaled Clusters'] == 2]

plt.scatter(df1.Age, df1['Income($)'], color = 'red')
plt.scatter(df2.Age, df2['Income($)'], color = 'blue')
plt.scatter(df3.Age, df3['Income($)'], color = 'green')
plt.scatter(S_Centroid[:,0], S_Centroid[:,1], color = 'purple', marker = '+', label = 'Centroid')

plt.title('KMeans Clustering: Scaled Age vs Scaled Income')
plt.xlabel('Age')
plt.ylabel('Income($)')

plt.legend()

plt.show()




# Step 10: Elbow Technique - To find the proper cluster with minimum error

KM_range = range(1,10)
SSE = []

for i in KM_range:

    KM = KMeans(n_clusters = i)
    KM.fit(df[['Age', 'Income($)']])
    SSE.append(KM.inertia_)


print('\n Sum of Squared Error :', SSE)


# Plot the Elbow Technique

plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(KM_range,SSE)
```

```python
# Example 2: Cluster iris flowers using their petal width and length

import os
# Fix memory leak warning on Windows with MKL and KMeans
os.environ["OMP_NUM_THREADS"] = "1"


import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler


# Step 1: Define the Dataset

Iris = load_iris()
print(dir(Iris))


df = pd.DataFrame(Iris.data, columns = Iris.feature_names)

df['Target'] = Iris.target
print('\n',df)



# Step 2: Initialize the variables

x = df.drop(['sepal length (cm)','sepal width (cm)'], axis = 'columns', inplace = True)
y = df['Target']



# Step 3: Assign the cluster & Train the Data using K Means

KM = KMeans(n_clusters = 3)
KM.fit(df)


# Step 4: Predict the Dataset

y_pred = KM.predict(df)
print('\n Y Predict :', y_pred)

df['Clusters'] = y_pred
print('\n',df)


# Centeroid point of the Dataset

Cenroid = KM.cluster_centers_
print('\n Cenroid :',Cenroid)

# Step 5: Plot the Dataset

df1 = df[df.Clusters == 0]
df2 = df[df.Clusters == 1]
df3 = df[df.Clusters == 2]

df1 = df[df['Clusters'] == 0]
df2 = df[df['Clusters'] == 1]
df3 = df[df['Clusters'] == 2]

plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color = 'red')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color = 'blue')
plt.scatter(df3['petal length (cm)'], df3['petal width (cm)'], color = 'green')
plt.scatter(Centroid[:,0], Centroid[:,1], color = 'purple', marker = '+', label = 'Centroid')

plt.title('KMeans Clustering: length vs width')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')

plt.legend()

plt.show()




# Step 6: Scale the Dataset - Use minmax scaler


Scaler = MinMaxScaler()

df['petal length (cm)'] = Scaler.fit_transform(df[['petal length (cm)']])
df['petal width (cm)'] = Scaler.fit_transform(df[['petal width (cm)']])

print('\n',df)



# Step 7: Train the Scaled Dataset

KM = KMeans(n_clusters = 3)
KM.fit(df)


# Step 8: Predict the Scaled Dataset

S_pred = KM.predict(df)
print('\n Scaled Predict :', y_pred)

df['Scaled Clusters'] = y_pred
print('\n',df)


# Centeroid point of the Scaled Dataset

Scaled_Centeroid = KM.cluster_centers_
print('\n Cenroid :',Scaled_Centeroid)


# Step 9: Plot the Scaled Dataset

df1 = df[df['Scaled Clusters'] == 0]
df2 = df[df['Scaled Clusters'] == 1]
df3 = df[df['Scaled Clusters'] == 2]


plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color = 'red')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color = 'blue')
plt.scatter(df3['petal length (cm)'], df3['petal width (cm)'], color = 'green')
plt.scatter(Scaled_Centeroid[:,0], Scaled_Centeroid[:,1], color = 'purple', marker = '+', label = 'Centroid')

plt.title('KMeans Clustering: Scaled length vs Scaled width')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')

plt.legend()

plt.show()




# Step 10: Elbow Technique

K_range = range(1,10)
SSE = []


for i in K_range:

    KM = KMeans(n_clusters = i)
    KM.fit(df)
    SSE.append(KM.inertia_)


print('\n Sum of Squared Error :', SSE)


# Plot the Elbow Technique

plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(KM_range,SSE)
    
```

# 2. Dimentionality Reduction

Dimensionality Reduction in unsupervised learning refers to the process of reducing the number of input variables or features in a dataset while preserving as much important information or structure as possible.

## 2.1. Principal Component Analysis (PCA)

PCA (Principal Component Analysis) is a popular dimensionality reduction technique used in unsupervised learning. It transforms the original high-dimensional data into a smaller set of uncorrelated variables called principal components, which capture the maximum variance (i.e., the most important patterns) in the data.



📌 Key Concepts:

Principal Components: New axes (directions) in the feature space that capture the most variance.

Orthogonal: All principal components are uncorrelated (perpendicular in space).

First PC: Captures the most variance.

Second PC: Captures the next highest variance, and so on.

```python
# Example 1: Handwritten digits dataset classification


import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA



# Step 1 : Define the Dataset


Digits = load_digits()
print(dir(Digits))

print('\n Digits Shape :',Digits.data.shape)


# To visualize Digits

print('\n Data 1 :', Digits.data[1])
print('\n Reshaped', Digits.data[1].reshape(8,8))

plt.gray()
plt.matshow(Digits.data[1].reshape(8,8))


# To know the target

Uni = np.unique(Digits.target)
print('\n Unique Target :', Uni)


# Create a Data Frame

df = pd.DataFrame(Digits.data, columns = Digits.feature_names)
print('\n Data Frame :',df.head())
print('\n Describe df :',df.describe())





# Step 2: Initialize the parameter

x = df
y = Digits.target


# Scale the features

SS = StandardScaler()
x_scaler = SS.fit_transform(x)
print('\n Scaled X :', x_scaler)




# Step 3: Split the Dataset  into train & test


x_train,x_test,y_train,y_test = train_test_split(x_scaler, y, test_size = 0.3)



# Step 4: Train the Dataset using LogisticRegression


LR = LogisticRegression()
LR.fit(x_train, y_train)

# Evaluate the model

Score = LR.score(x_test, y_test)*100
print('\n LR Score :', Score)



# Step 5: Reduce the dimension using PCA without the component

print('\n Original Shape :', x.shape)

pca = PCA(0.95)
x_pca = pca.fit_transform(x)

print('\n PCA Shape :', x_pca.shape)

print('\n PCA Ratio :', pca.explained_variance_ratio_)

print('\n PCA Components :', pca.n_components_)



# Step 6: Split the PCA Data into train & test

x_train_pca,x_test_pca,y_train,y_test = train_test_split(x_pca, y, test_size = 0.3)



# Step 7: Train the Dataset using LogisticRegression


pca_LR = LogisticRegression(max_iter = 1000)
pca_LR.fit(x_train_pca, y_train)

# Evaluate the model

pca_Score = pca_LR.score(x_test_pca, y_test)*100
print('\n PCA LR Score :', pca_Score)

```

```python
# PCA WITH COMPONENT


# Step 5: Reduce the dimension using PCA with the component

print('\n Original Shape :', x.shape)

pca = PCA(n_components = 2)
x_pca = pca.fit_transform(x)

print('\n PCA Shape :', x_pca.shape)

print('\n PCA Ratio :', pca.explained_variance_ratio_)

print('\n PCA Components :', pca.n_components_)



# Step 6: Split the PCA Data into train & test

x_train_pca,x_test_pca,y_train,y_test = train_test_split(x_pca, y, test_size = 0.3)



# Step 7: Train the Dataset using LogisticRegression


pca_LR = LogisticRegression(max_iter = 1000)
pca_LR.fit(x_train_pca, y_train)

# Evaluate the model

pca_Score = pca_LR.score(x_test_pca, y_test)*100
print('\n PCA LR Score :', pca_Score)
```

```python
# Example 2: Heart Disease

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA



# Step 1: Define the dataset

df = pd.read_csv(r"C:\Users\admin\Downloads\heart.csv")
print('Dataset : \n ',df.head())

print('\n Dataset Shape: ',df.shape)
print('\n Describe :',df.describe())


# Treat the Outliers

df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+3*df.Cholesterol.std())]
print('\n DF1 Cholesterol :',df1.shape)

df2 = df1[df1.Oldpeak<=(df1.Oldpeak.mean()+3*df1.Oldpeak.std())]
print('\n DF2 Oldpeak :',df2.shape)

df3 = df2[df2.RestingBP<=(df2.RestingBP.mean()+3*df2.RestingBP.std())]
print('\n DF3 RestingBP :',df3.shape)


# Unique

print('\n Unique ChestPainType :',df.ChestPainType.unique())
print('\n Unique RestingECG :',df.RestingECG.unique())
print('\n Unique ExerciseAngina :',df.ExerciseAngina.unique())
print('\n Unique ST_Slope :',df.ST_Slope.unique())


df4 = df3.copy()
pd.set_option('future.no_silent_downcasting', True)

# Replace categorical values

df4.replace({
    'ExerciseAngina': {'N': 0, 'Y': 1},
    'ST_Slope': {'Down': 0, 'Flat': 1, 'Up': 2},
    'RestingECG': {'Normal': 0, 'ST': 1, 'LVH': 2}
}, inplace=True)


print('\n DF4 :', df4.head())


df5 = pd.get_dummies(df4, drop_first=True)
print('\n DF5 :',df5.head())





# Step 2: Initialize the parameter


x = df5.drop("HeartDisease",axis='columns')
y = df5.HeartDisease


# Scale the features

SS = StandardScaler()
x_scaler = SS.fit_transform(x)
print('\n Scaled X :', x_scaler)



# Step 3: Split the PCA Data into train & test

x_train,x_test,y_train,y_test = train_test_split(x_scaler, y, test_size = 0.2, random_state=30)



# Step 4: Train the Dataset using Random Forest Classifier


RF = RandomForestClassifier()
RF.fit(x_train,y_train)

# Evaluate the model

Score = RF.score(x_test, y_test)*100
print('\n RF Score :', Score)



# Step 5: Reduce the dimension using PCA without the component

print('\n Original Shape :', x.shape)

pca = PCA(0.95)
x_pca = pca.fit_transform(x)

print('\n PCA Shape :', x_pca.shape)

print('\n PCA Ratio :', pca.explained_variance_ratio_)

print('\n PCA Components :', pca.n_components_)



# Step 6: Split the PCA Data into train & test

x_train_pca,x_test_pca,y_train,y_test = train_test_split(x_pca, y, test_size = 0.2, random_state=30)



# Step 7: Train the Dataset using LogisticRegression


pca_RF = RandomForestClassifier()
pca_RF.fit(x_train_pca,y_train)

# Evaluate the model

pca_Score = pca_RF.score(x_test_pca, y_test)*100
print('\n PCA RF Score :', pca_Score)
```
