# Performance Metrics

Performance metrics help you quantitatively evaluate how well your machine learning model is doing. The right metric depends on whether you're solving a classification, regression, or clustering problem.

## 1. Regression Metrics

### 1.1. MSE

MSE is a standard way to measure the error of a model in regression problems. It’s the average of the squared differences between the predicted values and the actual values.

🔍 Characteristics:

1. MSE is always non-negative.

2. Lower MSE = better model.

3. Because it squares the errors, it penalizes larger errors more than smaller ones.

4. It has the same units as the squared target variable.

```python
# Example:

import numpy as np

y_true = np.array([2.2, 2.8, 3.6, 4.5, 5.1])
y_pred = np.array([2.0, 2.9, 3.5, 4.4, 5.3])

mse = np.mean((y_true - y_pred) ** 2)
print("MSE:", mse)
```

### 1.2. RMSE 

RMSE stands for Root Mean Squared Error. It's one of the most common metrics used to evaluate the performance of regression models.



✅ What RMSE Tells You:

1. It measures the average magnitude of the error between predicted and actual values.

2. It penalizes large errors more than small ones (because of squaring).

3. Lower RMSE = better fit.

🧠 Why Use RMSE Instead of Just Error?

1. RMSE gives a sense of how spread out these residuals (errors) are.

2. Because the errors are squared, RMSE is more sensitive to outliers than MAE (Mean Absolute Error).

```python
# Example:

import numpy as np

y_true = np.array([2.2, 2.8, 3.6, 4.5, 5.1])
y_pred = np.array([2.0, 2.9, 3.5, 4.4, 5.3])

rmse = np.sqrt(np.mean((y_true - y_pred) ** 2))
print("RMSE:", rmse)
```

### 1.3. MAE

MAE stands for Mean Absolute Error — a simple and intuitive way to evaluate the accuracy of a regression model.

✅ What MAE Tells You:

1. It measures the average magnitude of the errors in a set of predictions.

2. It treats all errors equally (no squaring).

3. Lower MAE = better model performance.

```python
# Example:

import numpy as np

y_true = np.array([2.2, 2.8, 3.6, 4.5, 5.1])
y_pred = np.array([2.0, 2.9, 3.5, 4.4, 5.3])

mae = np.mean(np.abs(y_true - y_pred))
print("MAE:", mae)
```

### 1.4. R2 Score

The R² Score (R-squared), also known as the coefficient of determination, is a statistical measure that shows how well a regression model fits the actual data.

✅ What R² Tells You:

R² = 1 → perfect prediction

R² = 0 → model is no better than the mean

R² < 0 → model is worse than predicting the mean

It measures the proportion of the variance in the dependent variable that is predictable from the independent variable(s).

```python
# Example:

import numpy as np
from sklearn.metrics import r2_score

y_true = np.array([2.2, 2.8, 3.6, 4.5, 5.1])
y_pred = np.array([2.0, 2.9, 3.5, 4.4, 5.3])

# Method 1: Using sklearn
r2 = r2_score(y_true, y_pred)
print("R² Score (sklearn):", r2)

# Method 2: Manual calculation
ss_res = np.sum((y_true - y_pred)**2)
ss_tot = np.sum((y_true - np.mean(y_true))**2)
r2_manual = 1 - (ss_res / ss_tot)
print("R² Score (manual):", r2_manual)
```

## 2. Classification Metrics

### 2.1 Confusion Matrix

📊 What is a Confusion Matrix in Machine Learning?

A confusion matrix is a performance measurement tool for classification models, especially in binary and multiclass classification. It compares the actual target values with those predicted by the model.

🧠 Explanation of Terms:

TP (True Positive): Model predicted Yes, and it was Yes

TN (True Negative): Model predicted No, and it was No

FP (False Positive): Model predicted Yes, but it was No

FN (False Negative): Model predicted No, but it was Yes

```python

from sklearn.metrics import confusion_matrix

# Example actual and predicted values
y_true = [1, 0, 1, 1, 0, 1, 0]
y_pred = [1, 0, 1, 0, 0, 1, 1]

cm = confusion_matrix(y_true, y_pred)
print(cm)
```

🎯 Metrics Derived from Confusion Matrix:

Accuracy = (TP + TN) / Total

Precision = TP / (TP + FP)

Recall (Sensitivity) = TP / (TP + FN)

F1 Score = 2 * (Precision × Recall) / (Precision + Recall)

Output Means:

TN = 2

FP = 1

FN = 1

TP = 3

```python
# 📈 Visual Confusion Matrix:

import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
