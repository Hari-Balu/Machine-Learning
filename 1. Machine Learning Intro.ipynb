# MACHINE LEARNING

Machine Learning (ML) is a subset of artificial intelligence (AI) that enables computers to learn from data and improve their performance over time without being explicitly programmed.

In simple terms:

Machine learning is about teaching computers to make decisions or predictions based on past experiences (data), similar to how humans learn.

Example:

Suppose you want a system to recognize whether an email is spam or not.

1. Without ML: You would write manual rules (e.g., if the email has "buy now", mark as spam).

2. With ML: You feed it thousands of labeled emails (spam vs. not spam), and it learns patterns to classify new emails automatically.


### Main Types of Machine Learning:

1. Supervised Learning

🔹 Learn from labeled data (input → output).

🔹 Examples: Spam detection, predicting house prices.

2. Unsupervised Learning

🔹 Find patterns in unlabeled data.

🔹 Examples: Customer segmentation, topic modeling.

3. Reinforcement Learning

🔹 Learn by interacting with an environment and receiving feedback (rewards or penalties).

🔹 Examples: Game-playing AIs, robotics.

📊 Real-World Uses:

1. Netflix recommendations

2. Self-driving cars

3. Fraud detection

4. Voice assistants like Siri or Alexa

## 1. Supervised Learning

Supervised learning is a type of machine learning where the model is trained on a labelled dataset.

This means the algorithm learns from input-output pairs, so it can predict the output for new, unseen data.

🔁 Basic Concept:

✅ You provide the model with inputs (features) and the correct outputs (labels).

✅ The model learns a mapping function from inputs to outputs.

✅ Goal: Minimize the error between the model’s prediction and the actual label.

#### 🧰 Two Main Types of Supervised Learning:

📈 Regression

🏷️ Classification

#### 1.1.  📈 Regression

Regression is a type of supervised learning used to predict a continuous numeric value based on input data.

🧠 Key Idea:

The model learns the relationship between independent variables (features) and a dependent variable (target/output).

🔧 When to Use Regression:

Use regression when your target/output is a number (not a category).


📊 Example Use Cases:

1. Predicting house prices based on size, location, number of rooms

2. Estimating salary based on years of experience

3. Forecasting temperature for the next week

📘 Types of Regression Algorithms:

1. Linear Regression
   
2. Multivariate

3. Polynomial Regression
	    
4. Ridge/Lasso Regression	 


#### 1.2. 🏷️ Classification

Classification is a supervised learning technique where the goal is to predict a categorical label for input data.

🧠 Key Idea:

The model learns from labeled training data and assigns new inputs to one of two or more categories (classes).

🎯 Goal of Classification:

To learn a decision boundary that separates classes in the feature space.

🧮 Popular Classification Algorithms:

1. Logistic Regression (for binary problems)

2. Decision Trees

3. Random Forest

4. Support Vector Machines (SVM)

5. k-Nearest Neighbours (k-NN)

6. Neural Networks

## 2. Unsupervised Learning 

Unsupervised learning is a type of machine learning where the model is trained on unlabeled data — meaning there are no predefined outputs or categories.

The goal is to discover hidden patterns or structures in the data.


🔍 Key Characteristics:

1. No labeled output

2. The algorithm tries to find similarities, differences, or structures in the data on its own

3. Often used for exploration, grouping, and dimensionality reduction


🧰 Main Tasks in Unsupervised Learning:

1. 📦 Clustering
   
2. 🧭 Dimensionality Reduction
   

#### 2.1. 📦 Clustering

Clustering is an unsupervised learning technique that groups similar data points together based on patterns or features.

There are no labels — the algorithm decides the groups (clusters) on its own.

🧠 Common Clustering Algorithms:

1. K-Means: Divides data into k clusters by minimizing distance to the cluster center.

2. Hierarchical Clustering: Builds a tree of clusters by either merging or splitting.

3. DBSCAN: Groups data based on density, identifying outliers too.


#### 2.2. 🧭 Dimensionality Reduction

Dimensionality Reduction is the process of reducing the number of features (dimensions) in your dataset while preserving its core structure or information.

🔹 Why Reduce Dimensions?

1. To simplify data visualization (e.g., reducing 100 features to 2D for a scatter plot)

2. To improve model performance by removing noise and redundancy

3. To speed up computations

🧠 Common Algorithms:

1. PCA (Principal Component Analysis): Projects data to new axes that capture the most variance.

2. t-SNE: Useful for visualizing high-dimensional data in 2D or 3D while preserving local structure.

3. Autoencoders: Neural networks that learn compressed representations of data.

## 3. Reinforcement Learning (RL)

Reinforcement Learning is a type of machine learning where an agent learns to make decisions by interacting with an environment and receiving feedback in the form of rewards or penalties.

🧠 Core Idea:

🔹Agent: The learner or decision-maker

🔹Environment: Where the agent operates

🔹Action: What the agent does

🔹Reward: Feedback from the environment

🔹State: Current situation of the agent

🔹Policy: The strategy the agent follows to decide actions


🔁 Learning Loop:

1. The agent observes the state of the environment.

2. It takes an action based on a policy.

3. The environment gives a new state and a reward.

4. The agent learns to maximise long-term reward over time.


🎯 Goal:

To find a policy that maximizes cumulative reward over time (called return).


🕹️ Real-World Examples:

🔹 Game-playing (e.g., AlphaGo, Chess, Dota2 bots)

🔹 Robotics (walking, grasping, balancing)

🔹 Self-driving cars (lane-following, braking)

🔹 Dynamic pricing, online recommendation systems


🧠 Popular RL Algorithms:


1. Value-based: Q-Learning, Deep Q-Network (DQN)

2. Policy-based: REINFORCE, PPO (Proximal Policy Optimization)

3. Actor-Critic: A3C (Asynchronous Advantage Actor-Critic), DDPG

## 1. scikit learn

Scikit-learn is a powerful and widely-used Python library for machine learning. It provides simple and efficient tools for data mining and data analysis. Some of its key features include:

1. Supervised Learning:
Classification: Identifying which category an object belongs to (e.g., decision trees, random forests, logistic regression).

Regression: Predicting a continuous value (e.g., linear regression, support vector regression).

2. Unsupervised Learning:
Clustering: Grouping similar objects (e.g., K-means, DBSCAN).

Dimensionality Reduction: Reducing the number of features while maintaining data variance (e.g., PCA).

3. Model Selection:
Cross-validation, hyperparameter tuning, and grid search.

4. Preprocessing:
Scaling, normalizing, encoding categorical features, handling missing values.

5. Metrics:
Evaluating model performance using metrics like accuracy, precision, recall, F1-score, ROC-AUC, etc.

```python
# Example

from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
```

## 2. Matplotlib 

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for plotting and visualizing data in the form of graphs, charts, and other types of plots.

Key Features of Matplotlib:

🔹Basic Plots: Line plots, bar plots, histograms, scatter plots, etc.

🔹Customization: Control over various elements like titles, labels, axes, colors, and more.

🔹Multiple Plots: Create subplots for visualizing multiple charts together.

🔹Interactivity: Support for interactive plotting, zooming, and panning.

🔹Export: Save plots to different formats like PNG, PDF, SVG, etc.

```python
# Basic Example:

import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a plot
plt.plot(x, y)

# Adding title and labels
plt.title('Simple Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

### Types of Plots:

```python
# 1. Line Plot:

plt.plot(x,y)
plt.show()
```

```python
# 2. Bar Plot:

Categories = ['A', 'B', 'C', 'D']
Values = [1, 2, 3, 4]

plt.bar(Categories, Values)
plt.show()
```

```python
# 3. Histogram

Data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

plt.hist(Data, bins = 5)
plt.show()
```

```python
# 4. Scatter Plot:

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.scatter(x,y)
plt.show()
```

```python
# 5. Subplots:

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Plot 2")

plt.show()
```

### Customizing Plots:

```python
# Colors and Styles:

plt.plot(x,y, color = 'green', linestyle = '--', marker = 'o')

plt.show
```

```python
# Grid:

plt.plot(x,y)
plt.grid(True)

plt.show()
```

```python
# Legends:

plt.plot(x,y, label = 'x vs y')
plt.legend()

plt.show()
```

```python
# Saving Plots:

plt.plot(x,y)

plt.savefig("Plot.png")
```

#### 4. Heatmap: For visualizing correlation matrices or any other matrix-like data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('iris')

# Select only numeric columns
numeric_data = data.select_dtypes(include='number')

# Compute correlation matrix
correlation_matrix = numeric_data.corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

#### 5. Regressions:

```python
# Regression Plot (scatter plot with a regression line):

sns.regplot(x='sepal_length', y='sepal_width', data=data)
plt.show()
```

#### 6. FacetGrid: For creating multi-plot grids based on data categories.

```python

g = sns.FacetGrid(data, col="species")
g.map(sns.histplot, 'sepal_length')

plt.show()
```
