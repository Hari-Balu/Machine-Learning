# Feature Engineering

Feature Engineering is the process of transforming raw data into meaningful input features that improve the performance of machine learning models. It involves selecting, modifying, or creating new features from the original dataset to better capture the underlying patterns.

Why Feature Engineering is Important?

Machine learning algorithms rely heavily on the quality of the input features. Well-engineered features can:

1. Enhance model accuracy

2. Reduce training time

3. Improve interpretability

4. Help models generalize better to unseen data

## 1. Outlier detection and removal using percentile

Outlier detection and removal using percentiles is a simple and effective method based on identifying data points that fall outside a specified percentile range in a numerical feature.

 What Are Percentiles?

1. Percentiles indicate the relative standing of a value within a dataset.

2. For example, the 1st percentile is the value below which 1% of the data falls.

 Common Strategy for Outlier Removal

Typically, you keep data within a specific range, such as the 1st to 99th percentile or 5th to 95th percentile, and treat values outside this range as outliers.

```python
# Example 1: Height Outlier Detection

import pandas as pd

# Step 1: Dataset

df = pd.read_csv(r"C:\Users\admin\Downloads\heights.csv")
print(df)


# Step 2: Shape & Details of the Dataset

print('\n Shape :',df.shape)
print('\n Info :',df.describe())


# Step 3: Detect outliers using a percentile

Min_Threshold,Max_threshold = df.height.quantile([0.05,0.95])
print('\n',df[df.height < Min_Threshold])
print('\n',df[df.height > Max_threshold])


# Step 4: Remove outliers

df = df[(df.height > Min_Threshold) & (df.height < Max_threshold)]
print('\n',df)

```

## 2. Outlier detection and removal: Standard deviation

Outlier detection using the 3 standard deviations rule is based on the assumption that the data is normally distributed. It identifies as outliers any data points that fall more than 3 standard deviations away from the mean.

```python
# Example 1: Height Outlier Detection using standard deviation


import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np


# Step 1: Dataset

df = pd.read_csv(r"C:\Users\admin\Downloads\heights (1).csv")
print(df.head())

# Plot the Dataset

plt.hist(df.height, bins=20, rwidth=0.8)
plt.xlabel('Height (inches)')
plt.ylabel('Count')
plt.show()

plt.hist(df.height, bins=20, rwidth=0.8, density=True)
plt.xlabel('Height (inches)')
plt.ylabel('Count')

rng = np.arange(df.height.min(), df.height.max(), 0.1)
plt.plot(rng, norm.pdf(rng,df.height.mean(),df.height.std()))




# Step 2: Shape & Details of the Dataset

print('\n Shape :',df.shape)
print('\n Info :',df.describe())




# Step 3: Detect outliers using a Standard Deviation

Mean = df.height.mean()
print('\n Mean :',Mean)

Std_Dev = df.height.std()
print('\n Standard Deviation :',Std_Dev)


Upper_Limit = df.height.mean() + 3*df.height.std()
print('\n Upper_Limit :',Upper_Limit)

Lower_Limit = df.height.mean() - 3*df.height.std()
print('\n Lower_Limit :',Lower_Limit)


Outliers = df[(df.height > Upper_Limit) | (df.height < Lower_Limit)]
print('\n Outliers :',Outliers)




# Step 4: Remove outliers

df = df[(df.height < Upper_Limit) & (df.height > Lower_Limit)]
print('\n', df.sample(10))

print('\n Final Shape :',df.shape)
```

## 3. Outlier detection and removal: Z Score

 Outlier Detection Using Z-Score

The Z-Score method detects outliers by measuring how many standard deviations a data point is from the mean. It assumes the data is approximately normally distributed.

```python
# Example 1: Height Outlier Detection using Z-score


import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np


# Step 1: Dataset

df = pd.read_csv(r"C:\Users\admin\Downloads\heights (1).csv")
print(df.head())



# Step 2: Shape & Details of the Dataset

print('\n Shape :',df.shape)
print('\n Info :',df.describe())




# Step 3: Detect outliers using a Standard Deviation

Mean = df.height.mean()
print('\n Mean :',Mean)

Std_Dev = df.height.std()
print('\n Standard Deviation :',Std_Dev)


Z_score = (df.height - df.height.mean())/df.height.std()
df['Z Score'] = Z_score
print('\n',df.head())


Outliers = df[(df['Z Score'] < -3) | (df['Z Score'] > 3)]
print('\n Outliers :',Outliers)




# Step 4: Remove outliers

df = df[(df['Z Score'] > -3) & (df['Z Score'] < 3)]
print('\n', df.sample(20))

print('\n Final Shape :',df.shape)





```

## 4. Outlier detection and removal using IQR

 Outlier Detection Using IQR (Interquartile Range)

The IQR method is a robust way to detect outliers, especially for non-normally distributed data. It uses the range between the 1st and 3rd quartiles to determine what counts as an outlier.

```python
# Example 1: Height Outlier Detection using IQR


import pandas as pd



# Step 1: Defining the Dataset

df = pd.read_csv(r"C:\Users\admin\Downloads\heights (2).csv")
print(df.sample(10))

print('\n',df.describe())
print('\n Shape',df.shape)


# Step 2: Find the percentile

Q1 = df.height.quantile(0.25)
Q3 = df.height.quantile(0.75)

IQR = Q3 - Q1

print('\n Q1 :',Q1)
print('\n Q3 :',Q3)
print('\n IQR :',IQR)



# Step 3: Find the Outliers

Upper_Bound = Q3 + 1.5*IQR
print('\n Upper Bound :',Upper_Bound)

Lower_Bound = Q3 - 1.5*IQR
print('\n Lower Bound :',Lower_Bound)


Outliers = df[(df.height > Upper_Bound) | (df.height < Lower_Bound)]
print('\n Outliers :',Outliers)



# Step 4: Remove the Outliers

df = df[(df.height < Upper_Bound) & (df.height > Lower_Bound)]
print('\n',df.sample(10))
print('\n Final Shape',df.shape)

```
